```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

df=pd.read_csv("IPL.csv")
df.head()
```

```python
#some basic questions
```

```python
#1->which team win most matches
```

```python
match_wins=df['match_winner'].value_counts() #value count gibe series
sns.barplot(y=match_wins.index,x=match_wins.values,palette='viridis')
plt.title("Most Match win by team")
```

```python
#2-Toss Decision Trends
```

```python
sns.countplot(x=df['toss_decision'],palette='rainbow')
plt.title("Toss Decision Trends")
```

```python
#3-Toss winner vs Match Winner
```

```python
count=df[df['toss_winner']==df['match_winner']]['match_id'].count()
percentage=(count*100)/df.shape[0]
percentage.round(2)
```

```python
#How do teams wins (Runs vs wicket)
```

```python
sns.countplot(x=df['won_by'])
plt.title('Won by')
```

```python
#KEY PLAYER PERFORMACES
```

```python
#1-MOST 'PLAYER OF THE MATCH' AWARDS
```

```python
count=df['player_of_the_match'].value_counts().head(10)
count
```

```python
sns.barplot(x=count.values,y=count.index,palette='mako')
plt.title("Most Player of the Match")
```

```python
#2 Top Scorer
```

```python
high=df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
high
```

```python
high.plot(kind='barh')
```

```python
#10 BEST BOWLING FIGURES
df
```

```python
df['highest_wickets']=df['best_bowling_figure'].apply(lambda x: x.split('--')[0])
df['highest_wickets']=df['highest_wickets'].astype(int)
top_bowlers=df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowlers.plot(kind='barh')
```

```python
#VENUE ANALYSIS
```

```python
#1-MOST MATCHES PLAYED BY VENUE
```

```python
venue_count=df['venue'].value_counts()
venue_count
```

```python
sns.barplot(x=venue_count.values,y=venue_count.index,palette='rainbow')
plt.title("Venue Count")
```

```python
#CUSTOM QUESTIONS AND INSIGHTS
```

```python
#1-WHO WON HIGHEST MARGIN BY RUNS?
```

```python
df[df['won_by']=="Runs"].sort_values(by="margin",ascending=False).head(1)[['match_winner','margin']]
```

```python
#2 WHICH PLAYER HAD THE HIGHEST INDIVIDUAL SCORE?
```

```python
df[df['highscore']==df['highscore'].max()][['top_scorer','highscore']]
```

```python
#3 WHICH BOWLER HAD THE BEST BOWLING FIGURE?
```

```python
df[df['highest_wickets']==df['highest_wickets'].max()][['best_bowling','best_bowling_figure']]
```
