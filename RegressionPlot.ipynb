```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(
    x='total_bill',
    y='tip',
    hue='sex',
    data=tips,
    markers=['o', 'v'],
    palette='coolwarm',
    scatter_kws={'s':200}
)
```
