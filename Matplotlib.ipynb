```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x= np. linspace(0,5,11)
x
y=x**2
y
```

```python
#BASIC MATPLOTLIB
```

```python
plt.plot(x,y)
plt.title('Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
#SUBPLOT
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

```python
#OBJECT ORIENTED WAY
```

```python
#MAtplotlib allow to aspect ratio.DPI and figsize to be specified when the figure object is created.you can use the figsize and dpi keyword argument
#figsize is the tuple of width and height of the figure in inces.dpl is the dots-per-inch(pixel per inch)
```

```python
fig=plt.figure() #fig=plt.figure(figsize=(2,8),dpi=100)
axis1=fig.add_axes([0.1,0.1,1,1]) #[left side,bottom,height,width]
axis1.plot(x,y)
axis2=fig.add_axes([0.2,0.4,0.3,0.3]) #[left side,bottom,height,width]
axis2.plot(x,y)
```
